```python
# ruff: noqa
# type: ignore
# Ruff! Please spare me! I just need to plot the results...
import os

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


def add_line_to_plot(data_path, label=None):
    # import all csvs in the folder into each df
    files = os.listdir(data_path)

    dfs = []
    for file in files:
        if file.endswith(".csv"):
            df = pd.read_csv(os.path.join(data_path, file))
            dfs.append(df)

    # concatenate all dataframes into one
    df = pd.concat(dfs)

    # calculate median and standard deviation of the loss values
    median = df.groupby("Step")["Value"].median().reset_index()
    std = df.groupby("Step")["Value"].std().reset_index()
    upper_bound = median["Value"] + std["Value"]
    lower_bound = median["Value"] - std["Value"]

    # add the median loss line to the plot
    sns.lineplot(x="Step", y="Value", data=median, linewidth=2, label=label, alpha=0.8)
    plt.fill_between(median["Step"], upper_bound, lower_bound, alpha=0.2)
```

```python
# Reach
# create the plot
plt.figure(figsize=(5, 4))
plt.xlabel("Timesteps")
plt.ylabel("Mean Reward per Episode")
# plt.yscale('log')
plt.title("Reach")

# add lines to the plot
lines = [
    ("Baseline", "reach"),
    ("Autoencoder", "reach_ae"),
    ("MultiSegmenter", "reach_multiseg"),
]
for label, data_path in lines:
    add_line_to_plot("data/" + data_path, label=label)
# add_line_to_plot('data3', label='Data 3')

# show the plot
plt.legend()
# more grid lines in the background
plt.grid(which="both", axis="both", linestyle="--", alpha=0.5)
# nicer colors
plt.style.use("seaborn")

plt.savefig("figures/reach.pdf")
# plt.show()
```

```python
import os

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


def add_line_to_plot(data_path, label=None):
    # import all csvs in the folder into each df
    files = os.listdir(data_path)

    dfs = []
    for file in files:
        if file.endswith(".csv"):
            df = pd.read_csv(os.path.join(data_path, file))
            dfs.append(df)

    # concatenate all dataframes into one
    df = pd.concat(dfs)

    # calculate median and standard deviation of the loss values
    median = df.groupby("Step")["Value"].median().reset_index()
    std = df.groupby("Step")["Value"].std().reset_index()
    upper_bound = median["Value"] + std["Value"]
    lower_bound = median["Value"] - std["Value"]

    # add the median loss line to the plot
    sns.lineplot(x="Step", y="Value", data=median, linewidth=2, label=label, alpha=0.8)
    plt.fill_between(median["Step"], upper_bound, lower_bound, alpha=0.5)


# create the plot
plt.figure(figsize=(5, 4))
plt.xlabel("Training Steps")
plt.ylabel("EMD loss (MSE feature loss)")
plt.yscale("log")
plt.title("Autoencoder")

# add lines to the plot
lines = [("Table", "table_ae"), ("Peg In Hole", "pih_ae")]
for label, data_path in lines:
    add_line_to_plot("data/" + data_path, label=label)
# add_line_to_plot('data3', label='Data 3')

# show the plot
plt.legend()
# more grid lines in the background
plt.grid(which="both", axis="both", linestyle="--", alpha=0.5)
# nicer colors
plt.style.use("seaborn")

plt.savefig("figures/loss.pdf")
# plt.show()
```
